```python
import re
import requests
import time
from bs4 import BeautifulSoup
import os

import numpy as np
import pandas as pd
```

```python
#Defining the scraping function for BusinessWire
df = pd.DataFrame(columns=np.arange(6))
df.columns =['title', 'articleURL', 'time', 'channel', 'articleSummary', 'articleContent']

def bw_scraper(soup, df):
  df_new = pd.DataFrame(columns=np.arange(6))
  df_new.columns =['title', 'articleURL', 'time', 'channel', 'articleSummary', 'articleContent']
  for news in soup.find_all(itemtype="http://schema.org/NewsArticle"):
    news_url = news['itemid']
    res = requests.get(news_url, timeout=15)
    soup = BeautifulSoup(res.text, "html.parser")
    try:
      news_title = soup.find(itemprop="headline")['content']
    except:
      news_title = 'NA'
    try:
      news_subhead = soup.find(class_="bw-release-subhead").text
    except:
      news_subhead = 'NA'
    try:
      news_time = soup.find(class_="bw-release-timestamp").time['datetime']
    except:
      news_time = 'NA'
    try:
      news_content = [para.get_text() for para in soup.find(class_="bw-release-story").find_all('p')]
      news_content = ''.join(news_content)
    except:
      news_content = 'NA'
    time.sleep(1)
    news_channel = 'BusinessWire'
    df_new = df_new.append({
        'title': news_title,
        'articleURL': news_url, 
        'time': news_time, 
        'channel': news_channel, 
        'articleSummary': news_subhead, 
        'articleContent': news_content
    }, ignore_index=True)
  return df.append(df_new)
```
